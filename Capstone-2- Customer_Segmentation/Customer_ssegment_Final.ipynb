# 3. Customer Segmentation with RFM Scores

Businesses have this ever-lasting urge to understand their customers. ## Calculate RFM Scoring

The simplest way to create customer segments from an RFM model is by using **Quartiles**. We will assign a score from 1 to 4 to each category (Recency, Frequency, and Monetary) with 4 being the highest/best value. The final RFM score is calculated by combining all RFM values. For Customer Segmentation, you will use the df_rfm data set resulting from the RFM analysis.
<br>
<br>
**Note**: Data can be assigned into more groups for better granularity, but we will use 4 in this case.

1. Divide the df_rfm into quarters

```python
# Calculate quartiles for Recency, Frequency, and Monetary
quartiles = rfm_table[['Recency', 'Frequency', 'Monetary']].quantile(q=[0.25, 0.5, 0.75])
quartiles
```

### i. Creating the RFM Segmentation Table


1. Create two functions, one for Recency and one for Frequency and Monetary. For Recency, customers in the first quarter should be scored as 4, this represents the highest Recency value. Conversely, for Frequency and Monetary, customers in the last quarter should be scored as 4, representing the highest Frequency and Monetary values.

```python
# Create functions to assign scores
def recency_score(x):
    if x <= quartiles['Recency'][0.25]:
        return 4
    elif x <= quartiles['Recency'][0.5]:
        return 3
    elif x <= quartiles['Recency'][0.75]:
        return 2
    else:
        return 1

def frequency_monetary_score(x, column_name):
    if x <= quartiles[column_name][0.25]:
        return 1
    elif x <= quartiles[column_name][0.5]:
        return 2
    elif x <= quartiles[column_name][0.75]:
        return 3
    else:
        return 4
```

2. Score customers from 1 to 4 by applying the functions you have created. Also create separate score column for each value.

```python
# Apply the functions to calculate scores
rfm_table['Recency_Score'] = rfm_table['Recency'].apply(recency_score)
rfm_table['Frequency_Score'] = rfm_table['Frequency'].apply(frequency_monetary_score, column_name='Frequency')
rfm_table['Monetary_Score'] = rfm_table['Monetary'].apply(frequency_monetary_score, column_name='Monetary')

# Calculate RFM Score
rfm_table['RFM_Score'] = rfm_table['Recency_Score'] + rfm_table['Frequency_Score'] + rfm_table['Monetary_Score']

rfm_table.head()
```

3. Now that scored each customer, you'll combine the scores for segmentation.

```python
# Calculate RFM Score
rfm_table['RFM_Segment'] = rfm_table['Recency_Score'].astype(str) + rfm_table['Frequency_Score'].astype(str) + rfm_table['Monetary_Score'].astype(str)

rfm_table.head()
```

4. Define rfm_level function that tags customers by using RFM_Scrores and Create a new variable RFM_Level

```python
# Define the rfm_level function
def rfm_level(row):
    if row['RFM_Score'] >= 9:
        return 'Best Customers'
    elif row['RFM_Score'] >= 7:
        return 'Loyal Customers'
    elif row['RFM_Score'] >= 5:
        return 'Big Spenders'
    elif row['RFM_Score'] >= 4:
        return 'Almost Lost'
    elif row['RFM_Score'] >= 2:
        return 'Lost Customers'
    else:
        return 'Lost Cheap Customers'
```

```python
# Apply the function to create the RFM_Level variable
rfm_table['RFM_Level'] = rfm_table.apply(rfm_level, axis=1)

rfm_table.head()
```

5. Calculate average values for each RFM_Level, and return a size of each segment

```python
# Calculate average values for each RFM_Level
rfm_level_avg = rfm_table.groupby('RFM_Level').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean'
}).round(2) # taking the first 2 numbers after the point

# Return the size of each segment
rfm_level_counts = rfm_table['RFM_Level'].value_counts().reset_index()
rfm_level_counts.columns = ['RFM_Level', 'Count']
```

```python
# Display average values and segment sizes
print("Average Values for Each RFM_Level:")
print(rfm_level_avg)
print("\nSize of Each Segment:")
print(rfm_level_counts)
```

### i. Define the Optimal Number of Clusters

```python
# function to compute hopkins's statistic for the dataframe X
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
def hopkins(X, ratio=0.05):

    if not isinstance(X, np.ndarray):
      X=X.values  #convert dataframe to a numpy array
    sample_size = int(X.shape[0] * ratio) #0.05 (5%) based on paper by Lawson and Jures

    #a uniform random sample in the original data space
    X_uniform_random_sample = uniform(X.min(axis=0), X.max(axis=0) ,(sample_size , X.shape[1]))

    #a random sample of size sample_size from the original data X
    random_indices=sample(range(0, X.shape[0], 1), sample_size)
    X_sample = X[random_indices]

    #initialise unsupervised learner for implementing neighbor searches
    neigh = NearestNeighbors(n_neighbors=2)
    nbrs=neigh.fit(X)

    #u_distances = nearest neighbour distances from uniform random sample
    u_distances , u_indices = nbrs.kneighbors(X_uniform_random_sample , n_neighbors=2)
    u_distances = u_distances[: , 0] #distance to the first (nearest) neighbour

    #w_distances = nearest neighbour distances from a sample of points from original data X
    w_distances , w_indices = nbrs.kneighbors(X_sample , n_neighbors=2)
    #distance to the second nearest neighbour (as the first neighbour will be the point itself, with distance = 0)
    w_distances = w_distances[: , 1]

    u_sum = np.sum(u_distances)
    w_sum = np.sum(w_distances)

    #compute and return hopkins' statistic
    H = u_sum/ (u_sum + w_sum)
    return H
```

```python
import datetime as dt

def get_month(x):
    return dt.datetime.strftime(x, '%Y-%m')
```

Now we will use the function created above to convert all the invoice dates into respective month date format.

```python
# Svaing months only in Invoice Month
df_cohort['InvoiceMonth'] = df_cohort['InvoiceDate'].apply(get_month)

# Grouping Invoice Month by Customer ID since we want to apply cohort analysis on each unique customer
grouping = df_cohort.groupby('CustomerID')['InvoiceMonth']
df_cohort['CohortMonth'] = grouping.transform('min') # taking the first date
```

### ii. Calculating time offset in Months i.e. Cohort Index:
Calculating time offset for each transaction will allows us to report the metrics for each cohort in a comparable fashion.
First, you will create 4 variables that capture the integer value of years, months for Invoice and Cohort Date using the get_date_int() function which you'll create it below.

```python
def get_date_int(df, column):
    df[column] = pd.to_datetime(df[column])# since we have the date like YYYY-MM we need to convert it to dt
    
    # Extract year and month
    year = df[column].dt.year
    month = df[column].dt.month
    return year, month
```

You will use this function to extract the integer values for Invoice as well as Cohort Date in 3 separate series for each of the two columns

```python
# Get the integer values for Invoice and Cohort Date
invoice_year, invoice_month = get_date_int(df_cohort, 'InvoiceMonth')
cohort_year, cohort_month = get_date_int(df_cohort, 'CohortMonth')
```

Use the variables created above to calcualte the difference in Months and years and store them in cohort Index column.

```python
# Calculate the difference in months and store it in the 'CohortIndex' column
df_cohort['CohortIndex'] = (invoice_year - cohort_year) * 12 + (invoice_month - cohort_month)
```
